```python
# Importar librerías para tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd  # Pandas es una poderosa librería para manipulación y análisis de datos en Python.

# Importar librerías para web scraping y manipulación de datos
# -----------------------------------------------------------------------
from bs4 import BeautifulSoup
import requests

# Importar librerías para automatización de navegadores web con Selenium
# -----------------------------------------------------------------------
from selenium import webdriver  # Selenium es una herramienta para automatizar la interacción con navegadores web.
from webdriver_manager.chrome import ChromeDriverManager  # ChromeDriverManager gestiona la instalación del controlador de Chrome.
from selenium.webdriver.common.keys import Keys  # Keys es útil para simular eventos de teclado en Selenium.
from selenium.webdriver.support.ui import Select  # Select se utiliza para interactuar con elementos <select> en páginas web.

# Importar librerías para pausar la ejecución
# -----------------------------------------------------------------------
from time import sleep  # Sleep se utiliza para pausar la ejecución del programa por un número de segundos.

# Configuraciones
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None)  # Establece una opción de Pandas para mostrar todas las columnas de un DataFrame.
```

```python
# Inicializa el navegador Chrome
driver = webdriver.Chrome()

driver.maximize_window()

driver.get("https://www.casadellibro.com/")
print("Accede a la web")
sleep(2)

driver.find_element("css selector", "#onetrust-accept-btn-handler").click()
print("accepta cookies")

sleep(3)

driver.find_element("css selector", "#m-c > div:nth-child(2) > a").click()
print("entra en categoría ficción")

sleep(3)

lista_libro = []

for i in range(1,6):
    driver.find_element("css selector", f"#l-p-c > div:nth-child({i}) > div:nth-child(4) > a").click()
    print("entra en la página del libro")
    sleep(2)
    nombre = driver.find_element("css selector", "#t-p-f").text
    print(nombre)
    autor = driver.find_element("css selector", "#a-p-f > p > a").text
    print(autor)
    precio = driver.find_element("css selector", "#p-pf-f").text
    print(precio)
    editorial = driver.find_element("css selector", "#App > div.cma-components.container.d-grid.svelte-f3gp65 > cma-component:nth-child(1) > div > area-3 > p").text
    print(editorial)
    try:
        idioma = driver.find_element("css selector", "#App > div.cma-components.container.d-grid.svelte-f3gp65 > cma-component:nth-child(4) > div > div.mt-4.caracteristicas.svelte-id2jbc > h3:nth-child(4)").text.replace("Idioma:\n", "")
        print(idioma)
    except:
        idioma ="no encontrado"
        print(idioma)
    try: 
        num_paginas = driver.find_element("css selector", "#App > div.cma-components.container.d-grid.svelte-f3gp65 > cma-component:nth-child(4) > div > div.mt-4.caracteristicas.svelte-id2jbc > div:nth-child(5)").texttext.replace("Número de páginas:\n", "")
        print(num_paginas)
    except:
        num_paginas = "no encontrado"
        print(num_paginas)
    lista_libro.append({"Título":nombre, "Nombre autor":autor, "Precio": precio, "Editorial": editorial, "Idioma": idioma, "Num páginas": num_paginas})
    sleep(2)
    driver.back()

driver.close()
```
